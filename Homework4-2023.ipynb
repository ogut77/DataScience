```python
from sklearn import datasets

diabetes = datasets.load_diabetes(as_frame=True)
print(diabetes['DESCR'])
```

```python
X=diabetes['data']
X.info()
```

```python
y=diabetes['target']
y
```

```python
def evalmetric(y,ypred):
 from scipy.stats import pearsonr
 import numpy as np
 e = y - ypred 
 mse_f = np.mean(e**2)
 rmse_f = np.sqrt(mse_f)
 mae_f = np.mean(abs(e))
 mape_f = 100*np.mean(abs(e/y))
 crl, _ = pearsonr(y, ypred)
 r2_f = crl*crl
 print("MSE:", mse_f)
 print("RMSE:", rmse_f)
 print("MAE:",mae_f)
 print("MAPE:",mape_f)
 print("R-Squared:", round(r2_f, 4))
```

```python
#2.Split the data into train and test using %25 percent of data
from sklearn.model_selection import StratifiedKFold, train_test_split
from sklearn.neighbors import KNeighborsClassifier
from sklearn.preprocessing import StandardScaler
from sklearn.tree import DecisionTreeClassifier

X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.25, random_state=16
)
```

```python
from sklearn.tree import DecisionTreeRegressor
dtr = DecisionTreeRegressor()
dtr.fit(X_train,y_train) 

print("Train data")
y_pred2 = dtr.predict(X_train)
evalmetric(y_train, y_pred2)
print("Test data")
y_pred2 = dtr.predict(X_test)
evalmetric(y_test, y_pred2)
```

```python
from sklearn.ensemble import RandomForestRegressor
rfr = RandomForestRegressor()
rfr.fit(X_train,y_train)

print("Train data")
y_pred = rfr.predict(X_train)
evalmetric(y_train, y_pred)
print("Test data")
y_pred = rfr.predict(X_test)
evalmetric(y_test, y_pred)
```
